```python
import pandas as pd
import plotly as px
```

```python
df = pd.read_csv("extracted_cap1cutoff.csv")
```

```python
df
```

```python
colleges_ranked = df[
    (
        (df["course_name"].str.contains("Computer")) |
        (df["course_name"].str.contains("Artificial"))
    ) &
    (df["categdory_reservation"].str.startswith("G"))
    ].sort_values("last_rank").groupby("college_name").head(1)
ranking = colleges_ranked[
    (colleges_ranked["college_name"].str.contains("pune", case=False)) |
    (colleges_ranked["college_name"].str.contains("COEP")) |
    (colleges_ranked["college_name"].str.contains("avasari", case=False)) |
    (colleges_ranked["college_name"].str.contains("pvg", case=False)) |
    (colleges_ranked["college_name"].str.contains("ravet", case=False)) |
    (colleges_ranked["course_university"].str.contains("Pune"))
    ][:30]
ranking
```

```python
for entry in enumerate(ranking["college_name"].to_list()):
    print(*entry)
```

```python
for city in ["pune", "mumbai", "nagpur", "amravati", "thane"]:
    print(city, df.loc[df["college_name"].str.contains(city, case=False, na=False), "college_name"].nunique())
```
